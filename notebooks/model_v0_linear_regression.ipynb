```python
# Load and visualize data
import pandas as pd

df_raw = pd.read_csv('/content/FoodWasteRecyleDataWithGDP-Population2002-2023.csv')
df_raw
```

```python
df_filtered = df_raw[['Non-Retail Food Establishments','Food Shops','Supermarkets','Population (000)','GDP per Capita (S$)','Recycling Rate (%)','Total Food Waste (kt)']]
df_filtered
```

```python
# Let's try to divide data between test and training sets. we use cross validation to predict target variable - total food waste (kt)
from matplotlib import pyplot as plt
import seaborn as sns
correlation_matrix = df_filtered.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
print(correlation_matrix["Total Food Waste (kt)"].sort_values(ascending=False))
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Define the target variable and selected features for the models
target_column = "Total Food Waste (kt)"
selected_features = [
    'Non-Retail Food Establishments',
    'Food Shops',
    'Supermarkets',
    'Population (000)',
    'GDP per Capita (S$)',
    'Recycling Rate (%)'
]

# Extract features and target variable from the dataset
feature_multivariate = df_raw[selected_features]
feature_univariate = df_raw[["Population (000)"]]
target = df_raw[target_column]

# Initialize a multivariate linear regression model
model_multivariate = LinearRegression()
loo = LeaveOneOut()  # Leave-One-Out Cross-Validation instance
mse_scores_multivariate = []  # List to store MSE for each iteration
actual_multivariate = []  # Store actual target values
predicted_multivariate = []  # Store predicted values

# Perform LOOCV for multivariate model
for train_index, test_index in loo.split(feature_multivariate):
    X_train, X_test = feature_multivariate.iloc[train_index], feature_multivariate.iloc[test_index]
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]

    model_multivariate.fit(X_train, y_train)  # Train model
    y_pred = model_multivariate.predict(X_test)  # Predict on test sample

    mse = mean_squared_error(y_test, y_pred)  # Compute MSE
    mse_scores_multivariate.append(mse)
    actual_multivariate.append(y_test.values[0])
    predicted_multivariate.append(y_pred[0])

# Compute average MSE across all LOOCV iterations
average_mse_multivariate = np.mean(mse_scores_multivariate)
print(f"Multivariate LOOCV MSE: {average_mse_multivariate}")

# Initialize a univariate linear regression model
model_univariate = LinearRegression()
mse_scores_univariate = []  # List to store MSE for each iteration
actual_univariate = []  # Store actual target values
predicted_univariate = []  # Store predicted values

# Perform LOOCV for univariate model
for train_index, test_index in loo.split(feature_univariate):
    X_train, X_test = feature_univariate.iloc[train_index], feature_univariate.iloc[test_index]
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]

    model_univariate.fit(X_train, y_train)  # Train model
    y_pred = model_univariate.predict(X_test)  # Predict on test sample

    mse = mean_squared_error(y_test, y_pred)  # Compute MSE
    mse_scores_univariate.append(mse)
    actual_univariate.append(y_test.values[0])
    predicted_univariate.append(y_pred[0])

# Compute average MSE across all LOOCV iterations
average_mse_univariate = np.mean(mse_scores_univariate)
print(f'Univariate LOOCV MSE: {average_mse_univariate}')

# Visualization: Predicted vs. Actual Values
plt.figure(figsize=(12, 6))

# Multivariate model plot
plt.subplot(1, 2, 1)
plt.scatter(actual_multivariate, predicted_multivariate)
plt.plot([min(actual_multivariate), max(actual_multivariate)],
         [min(actual_multivariate), max(actual_multivariate)], color='red')
plt.xlabel("Actual Values (Total Food Waste (kt))")
plt.ylabel("Predicted Values")
plt.title("Multivariate: Actual vs. Predicted (Total Food Waste (kt))")

# Univariate model plot
plt.subplot(1, 2, 2)
plt.scatter(actual_univariate, predicted_univariate)
plt.plot([min(actual_univariate), max(actual_univariate)],
         [min(actual_univariate), max(actual_univariate)], color='red')
plt.xlabel("Actual Values (Total Food Waste (kt))")
plt.ylabel("Predicted Values")
plt.title("Univariate: Actual vs. Predicted (Total Food Waste (kt))")

plt.tight_layout()
plt.show()

# NOTE:
# Actual values differ in order because LOOCV iterations are in a different order.
```

```python
# Extract the model parameters
univariate_intercept = model_univariate.intercept_
univariate_coefficient = model_univariate.coef_[0]

print(f"Intercept: {univariate_intercept}")
print(f"Coefficient: {univariate_coefficient}")
```

```python
# Let us use the projected population data to predict the total waste generated
# Data source:
# https://www.worldometers.info/world-population/singapore-population/

projection_data ="""
Year,Population
2030,6031377
2035,6153233
2040,6194782
2045,6156188
2050,6081691
"""
import io
projection_df = pd.read_csv(io.StringIO(projection_data))
projection_df['Population (000)'] = projection_df['Population']/1000

print(projection_df)
```

```python
# We can generate synthetic data from

projection_df["Predicted Food Waste (kt)"] = (
    projection_df["Population (000)"] * univariate_coefficient + univariate_intercept
)

print(projection_df)
```
