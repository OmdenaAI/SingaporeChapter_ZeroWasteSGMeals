```python
from geopy.distance import geodesic as gd

# the least and farthest distance between a supermarket and an NGO

# origin = (30.172705, 31.526725)  # (latitude, longitude) don't confuse
# dist = (30.288281, 31.732326)

# print(geodesic(origin, dist).meters)  # 23576.805481751613
# print(geodesic(origin, dist).kilometers)  # 23.576805481751613
# print(geodesic(origin, dist).miles)  # 14.64994773134371

df['geo'] = list(zip(df['latitude'], df['longitude']))
df2['geo'] = list(zip(df2['Lat,N,11,10'], df2['Long,N,11,10']))
dist_df = pd.DataFrame(columns=['origin', 'destination', 'distance', 'unit'])
dist_list = []



for index1, row1 in df.iterrows():  # Iterate using index and row
  for index2, row2 in df2.iterrows():  # Iterate using index and row
      lat1, lon1 = row1['geo']
      lat2, lon2 = row2['geo']
      dist_list.append(gd((lat1, lon1), (lat2, lon2)).kilometers)
      dist_df = pd.concat([dist_df, pd.DataFrame({
          'origin': [row1['LIC_NAME']],
          'destination': [row2['name,C,100']],
          'distance': [gd((lat1, lon1), (lat2, lon2)).kilometers],
          'unit': ['km']
      })], ignore_index=True)

len(dist_list)
dist_df.head()
```
